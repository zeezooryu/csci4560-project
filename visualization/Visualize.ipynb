```python
import sys
import json
import csv 

data = {}
data_separated = {}
geojson = {}

## Step 1: Load the best results data set
with open('best_run_fixed.csv', 'r') as f:
  reader = csv.DictReader(f);
  for row in reader:
    county_names = str(row['names'])
    district = int(row['district'])

    if county_names not in data:
        data[county_names] = {}

    if county_names not in data_separated:
        data_separated[county_names] = {}
    
    if county_names == "Cobb2" :
      county_names = "Cobb"
      data_separated[county_names] = {
        'district': district
      }
      print (county_names)
      print (data_separated[county_names])
    if county_names == "Dekalb2":
      county_names = "DeKalb"
      data_separated[county_names] = {
        'district': district
      }
      print (county_names)
      print (data_separated[county_names])
    if county_names == "Fulton2":
      county_names = "Fulton"
      data_separated[county_names] = {
        'district': district
      }
      print (county_names)
      print (data_separated[county_names])
    if county_names == "Gwinnett2":
      county_names = "Gwinnett"
      data_separated[county_names] = {
        'district': district
      }
      print (county_names)
      print (data_separated[county_names])
    else:
      # data_separated[county_names] = {
      #   'district': district
      # }
      data[county_names] = {
        'district': district
      }

## Step 2: Load the shapes for the counties (features), lookup the county names
## values based on the county names (name10), and set it as part of the feature
## properties
with open('Counties_Georgia.json', 'r') as f:
  geojson = json.load(f)
  for feature in geojson['features']:
    featureProperties = feature['properties']
    name = str(featureProperties['NAME10'])
    featureData = data.get(name)
    for key in featureData.keys():
      featureProperties[key] = featureData[key]

## Step 4: To get geojson file for just the 4 separated counties
with open('Counties_Georgia.json', 'r') as f:
  geojson2 = json.load(f)
  for feature in geojson2['features']:
    featureProperties = feature['properties']
    name = str(featureProperties['NAME10'])
    featureData = data_separated.get(name)
    for key in featureData.keys():
      featureProperties[key] = featureData[key]

## Step 5: Save the augmented shapefile
with open('Counties_Results.geojson', 'w') as f:
  json.dump(geojson, f)

with open('Counties_Results_Separated.geojson', 'w') as f:
  json.dump(geojson2, f)
```
